```python
import numpy as np
from matplotlib import pyplot as plt

# EMA
loss_list = np.load('./loss_list_without.npy')[ : 160]
alpha = 0.2
ema_loss_list = loss_list.copy()
for i in range(1, loss_list.shape[0]):
    ema_loss_list[i] = alpha * loss_list[i] + (1 - alpha) * ema_loss_list[i - 1]
plt.figure()
plt.grid()
plt.rcParams['font.sans-serif'] = ['SimHei'] # 
plt.rcParams['figure.figsize'] = (5.8, 4.0)
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'summer'
plt.plot(loss_list, label = 'loss')
plt.plot(ema_loss_list, label = 'ema_loss')
plt.legend()
plt.savefig('loss.png', dpi = 300)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

# EMA
ema_acc_list = np.load('./acc_list.npy')[:160]
acc_list = np.load('./acc_list_without.npy')[:160]

plt.figure()
plt.grid()
plt.rcParams['font.sans-serif'] = ['SimHei'] # 
plt.rcParams['figure.figsize'] = (5.8, 4.0)
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['figure.dpi'] = 300
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'summer'
plt.plot(acc_list, label = 'acc')
plt.plot(ema_acc_list, label = 'ema_acc')
plt.legend()
plt.savefig('acc.png', dpi = 300)
plt.show()
```
